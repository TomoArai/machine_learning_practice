```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=200, noise=0.25, random_state=3)
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1, aspect='equal', xlim=(-2, 2), ylim=(-2, 2))
ax.scatter(X[y == 0, 0], X[y == 0, 1], marker="o")
ax.scatter(X[y == 1, 0], X[y == 1, 1], marker="x")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
gbrt = GradientBoostingClassifier(n_estimators=1000, random_state = 0)
gbrt.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(max_depth=1,random_state = 0)
gbrt.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(random_state = 0, learning_rate=0.01)
gbrt.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(n_estimators=10000, random_state = 0, learning_rate=0.01)
gbrt.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
import numpy as np
from matplotlib.colors import ListedColormap
cmap = ListedColormap(('lightskyblue', 'orange'))
```

```python
x1_min, x1_max = -2, 2
x2_min, x2_max = -2, 2
x1_mesh, x2_mesh = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                                   np.arange(x2_min, x2_max, 0.1))
np.array([x1_mesh.ravel(), x2_mesh.ravel()]).T
```

```python
z = gbrt.predict(np.array([x1_mesh.ravel(), x2_mesh.ravel()]).T)
```

```python
z = z.reshape(x1_mesh.shape)   
```

```python
fig = plt.figure(figsize=(5, 5))

plt.contourf(x1_mesh, x2_mesh, z, alpha=0.2, cmap=cmap)
plt.xlim(-2., 2.)
plt.ylim(-2, 2.)

ax.contourf(x1_mesh, x2_mesh, z, alpha=0.2)
plt.scatter(X_test[y_test == 0, 0], X_test[y_test == 0, 1], marker="o")
plt.scatter(X_test[y_test == 1, 0], X_test[y_test == 1, 1], marker="x")

plt.show()
```
