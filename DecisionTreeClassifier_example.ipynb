```python
# import librarys

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import mglearn
```

```python
# making a data set

X, y = make_moons(n_samples=200, noise=0.25, random_state=3)
```

```python
# visualization

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1, aspect='equal', xlim=(-2, 2), ylim=(-2, 2))
ax.scatter(X[y == 0, 0], X[y == 0, 1], marker="o")
ax.scatter(X[y == 1, 0], X[y == 1, 1], marker="x")
plt.show()
```

```python
# spliting into train data and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
# machine learning by DecisionTreeClassifier

clf=DecisionTreeClassifier()
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
# oprimizing hyper parameter

clf=DecisionTreeClassifier(max_depth = 1)
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
clf=DecisionTreeClassifier(max_depth = 2)
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
clf=DecisionTreeClassifier(max_depth = 5)
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
clf=DecisionTreeClassifier(max_depth = 10)
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
# oprimized hyper parameter (max_depth =2 is the best)

clf=DecisionTreeClassifier(max_depth = 2)
clf.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
# visualization

fig, axes = plt.subplots(figsize=(5,5))
mglearn.plots.plot_tree_partition(X_train, y_train, clf, ax=ax)

mglearn.plots.plot_2d_separator(clf, X_train, fill=True, alpha=.4)
ax.set_title("DecisionTreeClassifier")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```
